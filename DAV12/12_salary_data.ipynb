```r
# Load the required libraries
library(ggplot2)
library(dplyr)
```

```r
# Load the dataset
salary_data <- read.csv("/content/salary_data.csv")

# View the first few rows of the dataset
head(salary_data)
```

```r
# Summary statistics of the dataset
summary(salary_data)
```

```r
# Scatter plot with regression line
ggplot(salary_data, aes(x=YearsExperience, y=Salary)) +
  geom_point() +                       # Scatter plot
  geom_smooth(method="lm", se=FALSE) + # Regression line
  labs(title="Salary vs. Years of Experience", x="Years of Experience", y="Salary") +
  theme_minimal()
```
