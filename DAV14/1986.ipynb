```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
# Assuming 'summer.csv' is the name of your dataset file
df = pd.read_csv('summer.csv')

```

```python
# A. How many unique events do we have thus far, in Summer Olympics?
num_unique_events = df['Event'].nunique()
print("A. Number of unique events in Summer Olympics:", num_unique_events)
```

```python
# B. What is the number of Sports held per year in Summer Olypics, since 1896?
sports_per_year = df.groupby('Year')['Sport'].nunique()
plt.figure(figsize=(10, 6))
sns.lineplot(x=sports_per_year.index, y=sports_per_year.values, marker='o')
plt.title('Number of Sports Held per Year in Summer Olympics (1896-2012)')
plt.xlabel('Year')
plt.ylabel('Number of Sports')
plt.grid(True)
plt.show()
```

```python
# C. How is the gender participation in Summer Olympics, since 1896?
gender_participation = df.groupby(['Year', 'Sex']).size().unstack()
gender_participation.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Gender Participation in Summer Olympics (1896-2012)')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# D. Overall Gender Distribution for Summer Olympics
overall_gender_distribution = df['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(overall_gender_distribution, labels=overall_gender_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Overall Gender Distribution for Summer Olympics')
plt.axis('equal')
plt.show()
```

```python
# E. Top 10 Men Athletes - Gold Medals
top_10_men_athletes = df[df['Sex'] == 'M'][df['Medal'] == 'Gold'].groupby('Athlete').size().nlargest(10)
print("\nE. Top 10 Men Athletes - Gold Medals:\n", top_10_men_athletes)
```

```python
# F. Top 5 countries to win most Gold medals in 2012
top_5_countries_2012 = df[df['Year'] == 2012][df['Medal'] == 'Gold'].groupby('Country').size().nlargest(5)
print("\nF. Top 5 Countries to Win Most Gold Medals in 2012:\n", top_5_countries_2012)
```
