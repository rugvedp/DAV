```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cocacola_df = pd.read_excel('/content/CocaCola_Sales_Rawdata.xlsx')
```

```python
cocacola_df.head()
```

```python
cocacola_df.shape
```

```python
def spliting_data(row):
    data = row['Quarter']
    data = data.split('_')
    quater, year = data[0], data[1]
    return quater, year

cocacola_df[['quater', 'year']] = cocacola_df.apply(spliting_data, axis=1, result_type='expand')
```

```python
cocacola_df.head()
```

```python
cocacola_df = cocacola_df.iloc[:,1:]
```

```python
cocacola_df.head()
```

#Year on Year Sales

```python
index = cocacola_df.groupby(['year'])['Sales'].sum().index
sales_yoy = cocacola_df.groupby(['year'])['Sales'].sum().values

plt.title('Year on Year Sales Growth')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.bar(index,sales_yoy)
```

```python
index = cocacola_df.groupby(['year'])['Sales'].sum().index
sales_yoy = cocacola_df.groupby(['year'])['Sales'].sum().values

plt.title('Year on Year Sales Growth')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.plot(index,sales_yoy)
```

#Each Quarter Sales contribution

```python
data = cocacola_df.groupby(['year','quater']).sum().reset_index()

plt.title('Sales Contribution in each quater')
plt.xlabel('Year')
plt.ylabel('Sales')
sns.barplot(x='year', y='Sales', hue='quater', data=data)
```

# Maximum sales in Which quater

```python
import pandas as pd

max_sales_by_quarter = cocacola_df.groupby(['year', 'quater'])['Sales'].max().unstack()
max_sales_by_quarter['max_quarter'] = max_sales_by_quarter.idxmax(axis=1)
max_sales_by_quarter
```

```python
plt.figure(figsize=(8, 6))
max_sales_by_quarter['max_quarter'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Max Sales Quarter Distribution')
plt.ylabel('')
plt.show()
```

```python
# prompt: Using dataframe cocacola_df: syggest a graph using matplotlib,pyplot or seaborn

import matplotlib.pyplot as plt

# Create a line chart of Sales over time
plt.plot(cocacola_df['year'], cocacola_df['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Coca-Cola Sales Over Time')
plt.show()
```

#Applying Arima

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

train_data = cocacola_df[cocacola_df['year'] < '94']
test_data = cocacola_df[cocacola_df['year'] >= '94']

model = ARIMA(train_data['Sales'], order=(2, 1, 1))
model_fit = model.fit()

n_forecast = len(test_data)
forecast = model_fit.forecast(steps=n_forecast)

rmse = mean_squared_error(test_data['Sales'], forecast, squared=False)
print("RMSE:", rmse)

plt.figure(figsize=(12, 6))
plt.plot(train_data.index, train_data['Sales'], label='Training Data')
plt.plot(test_data.index, test_data['Sales'], label='Test Data')
plt.plot(test_data.index, forecast, label='Forecast', color='red')
plt.title('ARIMA Forecast for CocaCola Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
model_fit.summary()
```

```python
residual = pd.DataFrame(model_fit.resid)
residual.plot()
residual.plot(kind='kde')
```

#Arima Training

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np

# Prepare time series data
time_series_data = df.groupby('Date')['Statistics.Flights.Delayed'].sum().reset_index()

# Train-test split
train_data = time_series_data.iloc[:-12]  # Use all but the last 12 months for training
test_data = time_series_data.iloc[-12:]   # Use the last 12 months for testing

# Train ARIMA model
model = ARIMA(train_data.set_index('Date')['Statistics.Flights.Delayed'], order=(5,1,0))  # Adjust order as needed
model_fit = model.fit()

# Make predictions
forecast = model_fit.forecast(steps=len(test_data))

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test_data['Statistics.Flights.Delayed'], forecast))
print("RMSE:", rmse)

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(train_data['Date'], train_data['Statistics.Flights.Delayed'], label='Training Data')
plt.plot(test_data['Date'], test_data['Statistics.Flights.Delayed'], label='Test Data')
plt.plot(test_data['Date'], forecast, label='Forecast', color='red')
plt.title('ARIMA Forecast for Flight Delays')
plt.xlabel('Date')
plt.ylabel('Number of Delays')
plt.legend()
plt.show()
```

```python
print(model_fit.summary())
```

```python
residual = pd.DataFrame(model_fit.resid)
residual.plot()
residual.plot(kind='kde')
```
