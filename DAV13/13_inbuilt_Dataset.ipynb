```r
# Load the required libraries
library(ggplot2)
library(dplyr)
```

```r
# Load an inbuilt dataset (for example, iris dataset)
data(iris)
```

```r
# Histogram
ggplot(iris, aes(x=Sepal.Length)) +
  geom_histogram(binwidth=0.1, fill="skyblue", color="black") +
  labs(title="Histogram of Sepal Length", x="Sepal Length", y="Frequency") +
  theme_minimal()
```

```r
# Line chart
ggplot(iris, aes(x=1:nrow(iris), y=Petal.Length, group=Species, color=Species)) +
  geom_line() +
  labs(title="Line Chart of Petal Length", x="Index", y="Petal Length") +
  theme_minimal()
```

```r
# Pie chart
pie_data <- iris %>%
  group_by(Species) %>%
  summarise(count=n()) %>%
  mutate(percent = count/sum(count) * 100)

pie(pie_data$percent, labels=pie_data$Species, main="Pie Chart of Species Distribution")
```
